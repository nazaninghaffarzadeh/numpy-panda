```python
df.rename(columns={'Sex':'sex'})
```

```python
def computer_age(cols):
  Age=cols[0]
  Pclass=cols[1]
  if pd.isnull(Age):
    if Pclass==1:
      return 38
    elif Pclass==2:
      return 30
    else:
      return 25
  else :
    return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(computer_age,axis=1)
```
